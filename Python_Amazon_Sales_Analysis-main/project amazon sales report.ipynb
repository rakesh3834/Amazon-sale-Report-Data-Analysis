```python
# Plot Count of Cities By state
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Dikstribution of state')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 states Data
top_10_states=df['ship-state'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Dikstribution of state')
plt.xticks(rotation=90)
plt.show()
```

## Note:- From above Graph Most of the Buyers are from Maharshtra state
